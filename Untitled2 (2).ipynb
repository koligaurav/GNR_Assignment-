```python
import gradio as gr
import numpy as np
import rasterio
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.stats import multivariate_normal
import seaborn as sns
import pandas as pd

class SatelliteImageClassifier:
    def __init__(self, image_path):
        with rasterio.open(image_path) as src:
            self.image = src.read()
            self.transform = src.transform
            self.crs = src.crs

        self.image_2d = self.image.transpose(1, 2, 0).reshape(-1, self.image.shape[0])
        self.class_names = ['Water', 'Vegetation', 'Built Area']

    def sample_training_data(self, samples_per_class=100):
        np.random.seed(42)

        water_mask = (self.image_2d[:, 0] < np.percentile(self.image_2d[:, 0], 25))
        vegetation_mask = ((self.image_2d[:, 1] > np.percentile(self.image_2d[:, 1], 50)) &
                           (self.image_2d[:, 2] < np.percentile(self.image_2d[:, 2], 50)))
        built_mask = ((self.image_2d[:, 0] > np.percentile(self.image_2d[:, 0], 75)) &
                      (self.image_2d[:, 1] > np.percentile(self.image_2d[:, 1], 75)))

        water_samples = self.image_2d[water_mask][:samples_per_class]
        vegetation_samples = self.image_2d[vegetation_mask][:samples_per_class]
        built_samples = self.image_2d[built_mask][:samples_per_class]

        X = np.vstack([water_samples, vegetation_samples, built_samples])
        y = np.concatenate([
            np.zeros(samples_per_class),
            np.ones(samples_per_class),
            np.full(samples_per_class, 2)
        ])

        return X, y

    def maximum_likelihood_classification(self, X, y):
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        class_means = []
        class_covs = []
        for c in range(len(self.class_names)):
            class_data = X_scaled[y == c]
            class_means.append(np.mean(class_data, axis=0))
            cov = np.cov(class_data.T)
            regularized_cov = cov + 0.01 * np.eye(cov.shape[0])
            class_covs.append(regularized_cov)

        image_scaled = scaler.transform(self.image_2d)
        likelihoods = np.zeros((self.image_2d.shape[0], len(self.class_names)))
        for c in range(len(self.class_names)):
            mvn = multivariate_normal(mean=class_means[c], cov=class_covs[c])
            likelihoods[:, c] = mvn.pdf(image_scaled)

        classification = np.argmax(likelihoods, axis=1)
        return classification

    def minimum_distance_classification(self, X, y):
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        class_means = []
        for c in range(len(self.class_names)):
            class_data = X_scaled[y == c]
            class_means.append(np.mean(class_data, axis=0))

        image_scaled = scaler.transform(self.image_2d)
        distances = np.zeros((self.image_2d.shape[0], len(self.class_names)))
        for c in range(len(self.class_names)):
            distances[:, c] = np.linalg.norm(image_scaled - class_means[c], axis=1)

        classification = np.argmin(distances, axis=1)
        return classification

    def visualize_classification(self, classification):
        classified_image = classification.reshape(self.image.shape[1], self.image.shape[2])

        plt.figure(figsize=(10, 8))
        plt.imshow(classified_image, cmap='viridis')
        plt.title('Classification Result')
        plt.colorbar(ticks=range(len(self.class_names)),
                     label='Class',
                     format=plt.FuncFormatter(lambda val, loc: self.class_names[int(val)]))
        plt.tight_layout()
        plt.savefig("output.png")  # Save for Gradio to display
        plt.close()
        return "output.png"


def classify_image(image_path, algorithm):
    classifier = SatelliteImageClassifier(image_path)
    X, y = classifier.sample_training_data()

    if algorithm == "Maximum Likelihood":
        classification = classifier.maximum_likelihood_classification(X, y)
    elif algorithm == "Minimum Distance":
        classification = classifier.minimum_distance_classification(X, y)

    output_image = classifier.visualize_classification(classification)
    return output_image


# Gradio Interface
interface = gr.Interface(
    fn=classify_image,
    inputs=[
        gr.Textbox(label="Image Path", placeholder="Enter the path to the satellite image"),
        gr.Radio(["Maximum Likelihood", "Minimum Distance"], label="Classification Algorithm"),
    ],
    outputs=gr.Image(type="filepath", label="Classified Image"),  # Corrected type
    title="Satellite Image Classifier",
    description="Upload a satellite image and select a classification algorithm to see the results."
)

if __name__ == "__main__":
    interface.launch()
```
